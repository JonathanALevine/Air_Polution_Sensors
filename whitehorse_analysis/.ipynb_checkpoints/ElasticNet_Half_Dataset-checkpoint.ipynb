```python
import openpyxl
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor

from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
from sklearn.model_selection import StratifiedKFold

from matplotlib import pyplot as plt
import datetime
import math
import pandas as pd
from scipy import stats
import seaborn as sn

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

# Read in the dataframe
dataframe = pd.read_pickle('merged_dataset_elasped_time_hours_added.pkl')
dataframe = dataframe.dropna()

# Convert to numpy array
# Seperate the target 
# Remove the target from the DATA matrix
DATA = dataframe.to_numpy()
target = DATA[:,12].reshape(-1,1)
DATA = np.delete(DATA, 12, 1)
# Delete the dates
DATA = np.delete(DATA, 0, 1)
# Separate the elasped time (hours)
# Remove from the DATA matrix
hours = DATA[:,0]
DATA = np.delete(DATA, 0, 1)

# Display the dataframe
dataframe
```

```python
dataframe_new = dataframe.drop(columns=['reference_pm'])
corrMatrix = dataframe_new.corr()
plt.subplots(figsize=(10,10))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
DATA[:,0].reshape(-1,1)
```

```python
linear_regression = LinearRegression().fit(DATA[:,0].reshape(-1,1), target)
# Plot the values graph with the regression line
plt.figure(figsize=(25,20))
plt.plot(DATA[:,0], target, 'o')
plt.plot(DATA[:,0], 
         linear_regression.predict(DATA[:,0].reshape(-1,1)),
         label='y= {:0.4f}*x +({:0.4f})'.format(linear_regression.coef_[0][0], linear_regression.intercept_[0]),
         linewidth=3)
plt.xlabel('Predicted Concentration (ugm3)', fontsize=25)
plt.ylabel('Reference Concentration (ugm3)', fontsize=25)
plt.legend(loc='best', fontsize=25)
```

```python
metrics.r2_score(target, linear_regression.predict(DATA[:,0].reshape(-1,1)))
```

```python
# Split the data set into two halves, getting matrices a and b
A, B = np.split(np.delete(DATA, len(DATA)-1, 0), 2)
# Add the last odd row to b
B = np.vstack([B, DATA[:][len(DATA)-1]])
# Split the target
a, b = np.split(np.delete(target, len(target)-1, 0), 2)
b = np.vstack([b, target[:][len(target)-1]])
# Split the hours
a_hours, b_hours = np.split(np.delete(hours, len(hours)-1, 0), 2)
hours[:][len(hours)-1]
b_hours = np.append(b_hours, hours[:][len(hours)-1])
```

```python
# Simple Linear Regression method
linear_regression = LinearRegression().fit(A, a)
print("Test set score: {:0.6f}".format(linear_regression.score(A, a)))
print('MAE: {:0.4f}'.format(metrics.mean_absolute_error(a, linear_regression.predict(A))))
R2 = metrics.r2_score(a, linear_regression.predict(A))
adjusted_R2 = 1 - ((1-R2**2)*(np.size(A[:,0],0)-1))/(np.size(A[:,0],0) - np.size(A[:0],1) - 1)
print('R2: {:0.4f}'.format(R2))
print('adjusted-R2: {:0.4f}'.format(adjusted_R2))
print('Estimator feature strengths: {}'.format(linear_regression.coef_))
```
